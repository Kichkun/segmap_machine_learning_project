```python
import h5py
import random
```

```python
file = h5py.File('data_fancy.hdf5')
```

```python
# Merge data18
with h5py.File('data20.hdf5', 'r') as dataset:
    with h5py.File('data_fancy.hdf5', 'r') as data_fancy:
        start_index = 100
        others_cntr = 0
        building_cntr = 0

        keys = list(dataset.keys())
        random.shuffle(keys)

        for i in keys:
            if building_cntr != 8 and dataset[i]['label'][()] == 2:
                name = 'id' + str(start_index + others_cntr + building_cntr)
                h5py.h5o.copy(dataset.id, i.encode('utf-8'), data_fancy.id, name.encode('utf-8'))
                building_cntr += 1
            elif others_cntr != 8 and dataset[i]['label'][()] in (0, 3):
                name = 'id' + str(start_index + others_cntr + building_cntr)
                h5py.h5o.copy(dataset.id, i.encode('utf-8'), data_fancy.id, name.encode('utf-8'))
                others_cntr += 1
```

```python
# Check results
with h5py.File('data_fancy.hdf5', 'r') as rfile:
    print(rfile.keys())
```

```python
# Merge fancy_cars
with h5py.File('fancy_cars.hdf5', 'r+') as fancy_cars:
    with h5py.File('data_fancy.hdf5', 'r+') as data_fancy:
        for i in fancy_cars.keys():
            h5py.h5o.copy(fancy_cars.id, i.encode('utf-8'), data_fancy.id, i.encode('utf-8'))
```

```python
# Check results
with h5py.File('data_fancy.hdf5', 'r') as rfile:
    print(rfile.keys())
```

```python
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go
import plotly.tools as tools

init_notebook_mode(connected=True)
```

```python
def show(data):
    points = go.Scatter3d(x=data[:, 0], y=data[:, 1], z=data[:, 2],
                         mode='markers',
                         marker=dict(
                             size=1,
                             color=data[:, 2],
                             colorscale='Jet'
                         ))
    fig = go.Figure(data=[points])
    iplot(fig, show_link = False)
```

```python
with h5py.File('data_fancy.hdf5', 'r') as rfile:
    print(len(list(rfile.keys())))
    for i in rfile.keys():
        print(rfile[i]['label'][()])
        show(rfile[i]['num0'][:])
```
